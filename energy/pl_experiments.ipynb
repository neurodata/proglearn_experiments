```python
X, y = load_data("Boston", verbose = True)

X_train = X[0:n_train]
y_train = y[0:n_train]

for algo in algos:
    print("Computing rolling MAPE for %s." % algo['name'])
    algo['model'] = init_model(X_train, y_train, algo)
    compute_rolling_mape(X, y, n_train, stride, algo)
```
